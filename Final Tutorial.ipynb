## Data Collection

The first stage of the Data Science Life Cycle is data collection. To determine correlation between median houseshold income and birth rate, we must locate and standardize alll datasets related to our topic. [The Maryland Open Data Portal](https://opendata.maryland.gov/) (MODP) provides access to statewide statistical data national census information relating to the state of Maryland. Datasets from The Maryland Open Data Portal have been verificed by the state, but are subject to change with time.

Using the "Request" Python library, we are able to access the information given by the MODP's API's. Here, we can access annual median income, population, and birth rate statistics for each county in Maryland. To prevent data skewing, we have selected a sample range consistent with each dataset. By selecting data from 2010 - 2022 in each data range, we are able to determine correlation betweeen birth rate and annual income over 12-year period.

During this project, we used the programming language Python. The language's extensive libraries allowed for easy data collection, visualization, and representation.

```python
income_url = "https://opendata.maryland.gov/resource/bvk4-qsxs.csv"
population_url = "https://opendata.maryland.gov/resource/r7ky-rq9s.csv"
birthrate_url = "https://opendata.maryland.gov/resource/vavn-j725.csv"

# Function to download CSV from URL
def download_csv(url):
    response = requests.get(url)
    response.raise_for_status()  # Check that the request was successful
    return pd.read_csv(StringIO(response.text))

# Download and load the datasets
county_income = download_csv(income_url)
county_population = download_csv(population_url)
maryland_birthrate = download_csv(birthrate_url)
```

The "Pandas" Python library allows us to the next step in the Data Science Process: data preprocessing. In order to later manipulate the data and accurately represent the collected information, we must correct and remove all inconsistencies between the datasets. This standardization process allows us to create a consistent flow of information. 

Here we used pandas dataframes. These objects were able to convert the datasets (entered as Comma-Seperated Values) into tables that could be accessed and transformed.

Although the information provided by the Maryland Open Data Portal has had its accuracy verified, when the data was first accessed, it was severely untidy. Counties were represented as column variables, instead of having each row represent an instance of the dataset. This organizaton of data violates the "Tidy Data" convention.

```python
# Removing Duplicate Rows in Dataframes
median_household_income = county_income[22:]

# Dropping Unnecessary Data
median_household_income = median_household_income[median_household_income["data"] != "MOE"]
median_household_income = median_household_income.drop(["date_created", "data", "maryland"], axis=1)

maryland_birthrate = maryland_birthrate.drop(["date_created", "maryland"], axis=1)

county_population = county_population.drop_duplicates()
county_population = county_population.drop(["date_created", "maryland"], axis=1)
```

```python
# Standardizing Row Names
column_names = ["Year",  
"Allegany County", 
"Anne Arundel County", 
"Baltimore County", 
"Baltimore City",
"Calvert County",
"Caroline County",
"Carroll County",
"Cecil County",
"Charles County",
"Dorchester County",
"Frederick County",
"Garrett County",
"Harford County",
"Howard County",
"Kent County",
"Montgomery County",
"Prince George's County",
"Queen Anne's County",
"St. Mary's County",
"Somerset County",
"Talbot County",
"Washington County",
"Wicomico County",
"Worcester County"]

# Renaming Columns
median_household_income.columns = column_names
maryland_birthrate.columns = column_names
county_population.columns = column_names

# Melting Data and Representing Counties as Row Values
median_household_income = median_household_income.melt(id_vars=['Year'], var_name='County', value_name='Median Income')
maryland_birthrate = maryland_birthrate.melt(id_vars=['Year'], var_name='County', value_name='Total Births')
county_population =  county_population.melt(id_vars=['Year'], var_name='County', value_name='Annual Population')

# Combining Dataframes
annual_income_and_birthrate = pd.merge(median_household_income, maryland_birthrate, on=["Year", "County"])
annual_income_and_birthrate = pd.merge(annual_income_and_birthrate, county_population, on=["Year", "County"])

# Normalizing Number of Births by Population Size (Birth Rate per 1000 people)
annual_income_and_birthrate["Annual Birth Rate Per 1000"] = (maryland_birthrate["Total Births"] / county_population["Annual Population"])  * 1000
annual_income_and_birthrate.head()
```

## Hypothesis Testing

At this part of the Data Science Pipeline, we need to conduct a hypothesis test to validate the strength of our results. Up until this point we have looked at visual representations of our data trends and recognized both linear and non-linear patterns in each created graph. However, to determine if there is stastical significance in our produced results we must use a hypothesis test. Here, our null hypothesis would be there is no correlation between median household income and birth rate per county. Our alternative hypothesis would be there exists some correlation between the annual median household income and the birth rate per county. After representing our findings visually, we would hope to reject the null hypothesis due to the linear trend demonstrated.

We preformed linear regression on the entire data set. As a result, the p-values generated were incredibly low. However, this can be explained by one of the earlier graphs. In the graph "Linear Regression: Income vs Birthrate per 1,000 People", we see that there is no singular line that can be used to make predictions about the birthrate and median household income for each county. This is because each county has diverse population numbers, land areas, and a plethora of other factors that make each county's socio-economic status unique. However, performing linear regression analysis on each county individually may show the existance of a relationship between median houshold income and birth rate.

```python
# Holds the independent and dependent variables
X = annual_income_and_birthrate["Median Income"].values
y = annual_income_and_birthrate["Annual Birth Rate Per 1000"].values

# Add a constant to the independent variable (required for statsmodels regression)
X_with_const = smf.add_constant(X)

# Fit the linear regression model
model = smf.OLS(y, X_with_const).fit()

# Print the summary of the regression model
print(model.summary())

# Predict the values using the Linear Regression model
linreg_y_pred = model.predict(X_with_const)

# Calculate the MSE
linreg_mse = mean_squared_error(y, linreg_y_pred)

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(X, y, alpha=0.5, label='Data')
plt.plot(X, linreg_y_pred, color='blue', label='Linear Regression Fit')
plt.xlabel('Median Income ($)')
plt.ylabel('Annual Birth Rate Per 1000')
plt.title('Linear Regression Fit\nLinear Regression MSE: {:.5f}'.format(linreg_mse))
plt.legend()
plt.show()
```

```python
# Although we cant get a good linear regression line from teh whole data, we can see that each county's line is kinda linear. 

counties = annual_income_and_birthrate['County'].unique()

# Initialize a list to store p-values, coefficients, and MSEs
results = []

plt.figure(figsize=(12, 8))

# Plot for each county
for county in counties:
    county_data = annual_income_and_birthrate[annual_income_and_birthrate['County'] == county]
    
    # Scatter plot for the county
    plt.scatter(county_data['Median Income'], county_data['Annual Birth Rate Per 1000'], alpha= 0.5, label= f'{county}')
    
    # Calculate slope using numpy
    m, b = np.polyfit(county_data['Median Income'], county_data['Annual Birth Rate Per 1000'], 1)
    
    # Generate x and y values for the linear regression line
    x_values = np.linspace(county_data['Median Income'].min(), county_data['Median Income'].max(), 100)
    y_values = m * x_values + b
    
    # Plot the regression line
    plt.plot(x_values, y_values, linewidth=2)
    
    # Calculate p-values and coefficients using statsmodels without adding a constant term
    X = county_data['Median Income']
    y = county_data['Annual Birth Rate Per 1000']
    model = smf.OLS(y, X).fit()
    p_value = model.pvalues[0]  # Get the p-value for the slope
    coefficient = model.params[0]  # Get the coefficient for the slope
    
    # Predict values and calculate MSE
    y_pred = model.predict(X)
    mse = mean_squared_error(y, y_pred)
    
    # Annotate the p-value on the plot
    plt.text(x_values.mean(), y_values.mean(), f'p={p_value:.3f}', fontsize=9, color='black', ha='center')
    
    # Store the results
    results.append({
        'County': county,
        'Coefficient': coefficient,
        'P-value': p_value,
        'MSE': mse
    })

# Add titles and labels
plt.title('Correlation between Income and Birthrate by County (2010 - 2022)')
plt.xlabel('Median Income ($)')
plt.ylabel('Annual Birth Rate Per 1000')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)
plt.grid(True)
plt.show()

# Convert results to a DataFrame
results_df = pd.DataFrame(results)
```

After performing linear regression on each county individually, we notice that, although there doesn't exist a statewide relationship between median household income a birth rate,
there exists individual trends for each county. We notice that the previous graph "Correlation Between Income and Birth Rate by County (2010 - 2022)" shows a negative relationship between median household income. In a majority of counties, the more a household makes annually, the less likely they are to have children. Although this relationship is suprising at first, there are several societal factors that validate these results. 

It would not be wrong to assume that a houshold that makes more money on average would have more children, because they can afford it. However, this assumption overlooks several factors contributing to annual household income. It is likely that households with a higher median income, have members with strong educational backgrounds. On average, careers that require post-secondary and post-baccalaureate education pay more than careers that don't. Those that pursue these types of jobs prioritize their career over starting a family. Thus, opportunity cost could be one of many variables that lead to a decrease in birthrate as average income increases. 

This negative relationship in emphasized with Baltimore County's data points. This county starts with one of the highest birth during data analysis. However, as time progresses, Baltimore County's median household income increases. As Baltimore County's median household income increases, their annual birth rate decreases.

Although we obsevered linear trends bewtween median houshold income and birth rates through linear regression, there was no line that perfectly any county's data. To generate a model that better estimates a county's birthrate and median household income, we need to use a non-linear classification.

```python
# Print the results
print("Results:")
print(results_df.head())

# Print the significant results (values less than the significance level (alpha = 0.05))
significant_results = results_df[results_df['P-value'] < 0.05]
print("\nSignificant results:")
print(significant_results.head())
```

## K-Nearest Neighbors

We determined earlier that because the p-value for the linear regression model was much higher than our significance level (alpha = 0.05), we failed to reject the null hypothesis, thus substantiating the fact that the model cannot be fit by a linear regression line. As a result, we turned to the K-Nearest Neighbors (KNN) algorithm, which is better suited for capturing complex, non-linear relationships. KNN does not assume any specific form for the data distribution, making it flexible for various types of data, and it can capture non-linear decision boundaries, which is useful given the complex relationship between median income and birth rate. We fit the KNN regression model on the data with 5 nearest neighbors. Then, the regression model was used to predict the annual birth rate per 1000 based on the median income. Next, the Mean Squared Error (MSE) was calculated to assess the model's performance. Finally, KNN regression results were plotted to visualize the actual data points, representing the relationship between median income and annual birth rate per 1000, and the blue line, representing the predictions made by the KNN regression model. It captures the non-linear trends in the data that a linear model could not.

The KNN regression model, which had an MSE of approximately 1.9504, provided a better fit compared to the linear regression model, which had a higher MSE of 2.55316. This indicates that KNN is more effective at capturing the non-linear relationship between median income and annual birth rate per 1000. The graph and the lower MSE of the KNN model support this conclusion. KNN's ability to capture non-linear relationships makes it a suitable choice for this dataset. 

```python
# Holds the independent and dependent variables
X = annual_income_and_birthrate['Median Income'].values.reshape(-1, 1)
y = annual_income_and_birthrate['Annual Birth Rate Per 1000'].values

# Fit the KNN model
knn_reg = KNeighborsRegressor(n_neighbors= 5)
knn_reg.fit(X, y)

# Predict the values using the KNN model
KNN_y_pred = knn_reg.predict(X)

# Calculate MSE
knn_mse = mean_squared_error(y, KNN_y_pred)

# Plot the results
plt.figure(figsize=(15, 10))
plt.scatter(X, y, alpha=0.5, label='Data')
plt.plot(X, KNN_y_pred, color='blue', label='KNN Regression Fit')
plt.title('KNN Regression\nMSE: {:.5f}'.format(knn_mse))
plt.xlabel('Median Income ($)')
plt.ylabel('Annual Birth Rate Per 1000')
plt.legend()
plt.show()
```
